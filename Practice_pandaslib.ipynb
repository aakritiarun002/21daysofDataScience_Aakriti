```python
test_dict = {'a':300, 'b':400, 'c':500}
pd.Series(data = test_dict) #creating dict in pandas
```

```python
#creating list using DataFrame
```

```python
test_data = [['Tommy Boi',100], ['Leo', 200], ['Akanksha',30000]]
df = pd.DataFrame(test_data,columns = ['Name','Age'])
df
```

```python
test_dict = {'Name':['Tommy Boi', 'Leo', 'Akanksha'], 'Age':[100,200,30000]}
df = pd.DataFrame(test_dict)
df
```
